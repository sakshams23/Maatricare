```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
X = df_clean[[
    'Age',
    'BodyTemp',
    'HeartRate',
    'SystolicBP',
    'DiastolicBP',
    'BMI',
    'HbA1c',
    'FastingGlucose'
]]

y = df_clean['Outcome']

# Encode the target labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Initialize model
model = RandomForestClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=le.classes_))

# Example: Predict on new single data point (replace with your values)
new_data = [[30, 98.6, 80, 120, 80, 22.0, 40, 5.5]]  # [Age, BodyTemp, HeartRate, SystolicBP, DiastolicBP, BMI, HbA1c, FastingGlucose]
predicted_class = le.inverse_transform(model.predict(new_data))[0]
print("Predicted risk level:", predicted_class)
```
